```python
x1= dt.drop(['Cards','Education','Gender','Student','Married'], axis=1)
```

```python
y=x1.iloc[:,-1]
```

```python
x=x1.iloc[:,0:5]
```

```python
x
```

```python
Q1 = x.quantile(0.25)
Q3 = x.quantile(0.75)
IQR = Q3-Q1
print (IQR)
```

```python
Q1
```

```python
Q3
```

```python
Q3+1.5*Q3
```

```python
Q1-1.5*Q3
```

```python
x = x[~((x<(Q1-1.5*IQR))
                       | (x>(Q3+1.5*IQR))).any(axis=1)]
```

```python
x["Ethnicity"] = x["Ethnicity"].astype('category')
x["Ethnicity"] = x["Ethnicity"].cat.codes
```

```python
x
```

```python
x = pd.get_dummies(x, columns=['Ethnicity'], drop_first=True)
```

```python
x
```

```python
len(x)
```

```python
len(y)
```

```python
final = pd.concat([x,y],axis=1,join='inner')
```

```python
final
```

```python
y = final.iloc[:,-1].values
```

```python
y
```

```python
len(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm = lm.fit(x_train,y_train)
```

```python
y_pred=lm.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
from sklearn.tree import DecisionTreeRegressor
lm  = DecisionTreeRegressor()
lm.fit(x_train,y_train)
```

```python
y_pred = lm.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=5)
regressor.fit(x_train,y_train)
```
